```python
import nibabel as nib
import numpy as np
from pathlib import Path

import matplotlib.pyplot as plt

INPUT_DIR = Path("/home/tibia/Documents/Brain_GPT_Dataset")
SLICE_OUTPUT_DIR = INPUT_DIR / "slices_png"
SLICE_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# Cherche tous les .nii.gz du dossier
nii_files = list(INPUT_DIR.glob("*.nii.gz"))

for file in nii_files:
    img = nib.load(str(file))
    data = img.get_fdata()  # shape (H, W, 24)
    base_name = file.stem  # sans extension

    # Vérification
    if data.ndim != 3 or data.shape[2] != 24:
        print(f"⚠️ {file.name} ignoré (shape {data.shape})")
        continue

    # Sauvegarder les slices
    for i in range(24):
        slice_2d = data[:, :, i]
        # Normalisation simple [0, 255]
        slice_norm = 255 * (slice_2d - np.min(slice_2d)) / (np.ptp(slice_2d) + 1e-8)
        slice_uint8 = slice_norm.astype(np.uint8)

        # Nom du fichier
        out_path = SLICE_OUTPUT_DIR / f"{base_name}_slice{i:02d}.png"
        plt.imsave(out_path, slice_uint8, cmap="gray")

    print(f" {file.name} : 24 slices sauvegardées dans {SLICE_OUTPUT_DIR}")
```

```python
import os
import json

# chemins
data_dir = "/home/tibia/Documents/BrainGPT/FORTE/data"
ins_json = os.path.join(data_dir, "CQ500p_instruction.json")
output_json = os.path.join(data_dir, "CQ500p.json")

# charger les instructions
with open(ins_json, "r") as f:
    instructions = json.load(f)

mapping = {}

# parcourir chaque instruction
for ins_id, ins_data in instructions["data"].items():
    for image_id in ins_data["image_ids"]:
        fname = f"{image_id}.png"  # ex: ID_xxx_slice00.png
        fpath = os.path.join(data_dir, fname)

        if not os.path.exists(fpath):
            print(f"⚠️ Missing file: {fpath}")  # Alerte si un slice manque
        else:
            mapping[image_id] = fpath

# sauvegarder le mapping complet
with open(output_json, "w") as f:
    json.dump(mapping, f, indent=2)

print(f" CQ500p.json généré avec {len(mapping)} entrées")
```
